```python
class ResBlock(nn.Module):
    def __init__(self, dilation=1):
        super().__init__()
        
        self.dilation = dilation
        
        self.conv1 = nn.Conv1d(16, 16, 3, dilation=self.dilation, padding=self.dilation)
        self.batch1 = nn.BatchNorm1d(16)
        self.act1 = nn.ReLU()
        
        self.conv2 = nn.Conv1d(16, 16, 3, dilation=self.dilation, padding=self.dilation)
        self.batch2 = nn.BatchNorm1d(16)
        self.act2 = nn.ReLU()
        
        self.dropout = nn.Dropout(p=0.2)

    def forward(self, x):
        inp = self.dropout(x)
        dil1 = self.conv1(x)
        batch1 = self.batch1(dil1)
        act1 = self.act1(batch1)
        dil2 = self.conv2(act1)
        batch2 = self.batch2(dil2)
        act2 = self.act2(batch2)
        
        return inp + act2
```

```python
class Net1D(nn.Module):
    def __init__(self, n_out=0):
        super().__init__()
        
        self.n_out = n_out
        
        self.conv1 = nn.Conv1d(3, 16, 3)#, dilation=2)
        self.act1 = nn.ReLU()
        
        self.res1 = ResBlock(dilation=1)
        self.pool1 = nn.MaxPool1d(1, 2)

        self.res2 = ResBlock(dilation=1) 
        self.pool2 = nn.MaxPool1d(1, 2)

        self.res3 = ResBlock()
        
        self.fc = nn.Linear(256, self.n_out)

    def forward(self, x):
        # Fast.ai adds a fake dimension, remove it
        x = x.squeeze(1)
        
        x = self.conv1(x)
        x = self.act1(x)
        x = self.res1(x)
        x = self.pool1(x)
        x = self.res2(x)
        x = self.pool2(x)
        x = self.res3(x)
        x = torch.flatten(x, start_dim=1)
        x = self.fc(x)
        x = torch.Tensor(x.cpu().float()).float().cuda()
        
        return x
```
